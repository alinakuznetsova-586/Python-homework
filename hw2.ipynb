# Задание 1. Считайте датасет из файла train.csv

```python
import pandas as pd

df = pd.read_csv("train.csv")

print("Первые 5 строк:")
print(df.head())
```

# Задание 2. Выведите основную информацию о датасете: информацию о типах данных, число пропусков, средние значения и т.д.

```python
print("\nИнформация о датасете. Типы данных:")
print(df.info())

print("\nЧисло пропусков в каждой колонке:")
print(df.isnull().sum())

print("\nСтатистика числовых данных:")
print(df.describe())
```

# Задание 3. Посчитайте процент выживаемости у каждого класса пассажиров (Pclass)

```python
print("\nПроцент выживаемости у каждого класса:")
survival_rate = df.groupby("Pclass")["Survived"].mean() * 100
print(survival_rate)
```

# Задание 4. Выведите самое популярное мужское и самое популярное женское имя на корабле

```python
def get_first_name(name):
    if "(" in name:
        return name.split("(")[1].split(")")[0].split()[0]
    return name.split(".")[1].strip().split()[0]

names_df = df[["Sex", "Pclass"]].copy()
names_df["FirstName"] = df["Name"].apply(get_first_name)

print("\nСамые популярные имена:")

male_counts = (
    names_df[names_df["Sex"] == "male"]
    .groupby("FirstName")
    .size()
    .sort_values(ascending=False)
)

female_counts = (
    names_df[names_df["Sex"] == "female"]
    .groupby("FirstName")
    .size()
    .sort_values(ascending=False)
)

print(f"Мужское имя: {male_counts.index[0]} ({male_counts.iloc[0]})")
print(f"Женское имя: {female_counts.index[0]} ({female_counts.iloc[0]})")
```

# Задание 5. Выведите самое популярное мужское и самое популярное женское имя на корабле в каждом классе

```python
print("\nПопулярные имена по классам:")

for pclass in sorted(names_df["Pclass"].unique()):

    male_counts = (
        names_df[(names_df["Pclass"] == pclass) & (names_df["Sex"] == "male")]
        .groupby("FirstName")
        .size()
        .sort_values(ascending=False)
    )

    female_counts = (
        names_df[(names_df["Pclass"] == pclass) & (names_df["Sex"] == "female")]
        .groupby("FirstName")
        .size()
        .sort_values(ascending=False)
    )

    print(f"\nКласс {pclass}:")
    print(f"  Мужское: {male_counts.index[0]} ({male_counts.iloc[0]})")
    print(f"  Женское: {female_counts.index[0]} ({female_counts.iloc[0]})")
```

# Задание 6. Выведите часть таблицы с пассажирами, возраст которых больше 44 лет

```python
print("\nПассажиры старше 44 лет:")
print(df.sort_values(by="Age")[df["Age"] > 44][["Name", "Age"]])
```

# Задание 7. Выведите часть таблицы с пассажирами, возраст которых меньше 44 лет и которые мужского пола

```python
print("\nМужчины младше 44 лет:")
print(df.sort_values(by="Age")[(df["Age"] < 44) & (df["Sex"] == "male")][["Name", "Age"]])
```

# Задание 8. Выведите количества n-местных кабин (в которых было 2, 3, 4, ... человека)

```python
print("\nКоличество n-местных кабин:")

cabins_df = df[["Cabin"]].dropna().copy()
cabin_counts = cabins_df.groupby("Cabin").size()
n_person_cabins = cabin_counts.value_counts().sort_index()

print(n_person_cabins)
```
