# **Задание №1.**
## Работа со строками и исключениями


```python
def parse_int(s):
    strip_s = s.strip()
    try:
        result = int(strip_s)
        return result
    except ValueError:
        return None

s1 = "12"
s2 = " -7 "
s3 = "12a"
s4 = ""

print("Ввод:", s1, "Результат:", parse_int(s1))
print("Ввод:", s2, "Результат:", parse_int(s2))
print("Ввод:", s3, "Результат:", parse_int(s3))
print("Ввод:", s4, "Результат:", parse_int(s4))
```

# **Задание №2.**
## Циклы и работа со списками

```python
def count_positive(nums):
    count = 0
    for n in nums:
        if n > 0:
            count = count + 1
    return count

def sum_even(nums):
    summa = 0
    for n in nums:
        if n % 2 == 0:
            summa = summa + n
    return summa

def first_divisible(nums, k):
    for n in nums:
        if n % k == 0:
            return n
    return None

numbers = [10, -5, 3, 8, -2, 15, 0, 7, 22, -11, 4, 18, -30, 45, 1, 9, -6, 14, 2, 5]

print("Мой список чисел:", numbers)

cp = count_positive(numbers)
print("Положительных чисел:", cp)

se = sum_even(numbers)
print("Сумма чётных чисел:", se)

k = 3
fd = first_divisible(numbers, k)
print("Первое число, которое делится на", k, "- это", fd)
```

# **Задание №3.**
## Lambda и сортировки

```python
items = [
  {"name": "A", "price": 120, "rating": 4.7},
  {"name": "B", "price": 90, "rating": 4.2},
  {"name": "C", "price": 150, "rating": 4.9},
]

center = sorted(items, key=lambda x: x["price"])

rating = sorted(items, key=lambda x: x["rating"], reverse=True)

print("Сортировка по цене:")
for i in center:
    print(i)

print("Сортировка по рейтингу:")
for i in rating:
    print(i)
```

# **Задание №4.**
## Мини-ООП

```python
class Chocolate:
    def __init__(self, name, price):
        if price <= 0:
            raise ValueError("Цена должна быть больше нуля")
        self.name = name
        self.price = price

    def discount(self, percent):
        discount = self.price * percent / 100
        res = self.price - discount
        return res

choc1 = Chocolate("Шоколадка 1", 100.0)
choc2 = Chocolate("Шоколадка 2", 250.0)

print("Шоколад:", choc1.name, "Цена без скидки:", choc1.price)
print("Цена со скидкой 20%:", choc1.discount(20))

print("Шоколад:", choc2.name, "Цена без скидки:", choc2.price)
print("Цена со скидкой 15%:", choc2.discount(15))
```
