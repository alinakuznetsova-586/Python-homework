# Задание 1. Считайте датасет из файла train.csv (это данные о выживаемости на Титанике)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

df = pd.read_csv("train.csv")
df.head(15)
```

# Задание 2. Визуализируйте распределение значений признаков Survived, Pclass, Age, Sex, Parch

```python
plt.figure(figsize=(6, 4))
sns.countplot(x=df["Survived"])
plt.title("Распределение по колонке 'Survived'")
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(x=df["Pclass"])
plt.title("Распределение по колонке 'Pclass'")
plt.show()

plt.figure(figsize=(6, 4))
sns.histplot(df["Age"])
plt.title("Распределение по колонке 'Age'")
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(x=df["Sex"])
plt.title("Распределение по колонке 'Sex'")
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(x=df["Parch"])
plt.title("Распределение по колонке 'Parch'")
plt.show()
```

# Заданеи 3. Постройте график типа boxplot для столбца Age

```python
plt.figure(figsize=(6,4))

sns.boxplot(x=df["Age"])

plt.title("Boxplot для возраста пассажиров")
plt.xlabel("Age")

plt.show()
```

# Задание 4. Проинтерпретируйте полученный график

На графике видно, что больше всего пассажиров были в возрасте примерно от 20 до 40 лет. Средний возрастной диапазон сосредоточен около 30 лет. Также на графике есть несколько пассажиров пожилого возраста - примерно от 60 до 80 лет, и совсем маленькие дети, включая младенцев.

# Задание 5. Постройте график типа pie chart для переменных Survived, Pclass, подпишите доли в процентах

```python
survived_counts = df["Survived"].value_counts()
pclass_counts = df["Pclass"].value_counts().sort_index()

plt.figure(figsize=(6, 6))
survived_counts.plot(
    kind="pie",
    autopct="%1.1f%%",
    labels=["Не выжили", "Выжили"]
)
plt.title("Статистика выживаемости пассажиров на графике типа pie chart")
plt.ylabel("")
plt.show()

plt.figure(figsize=(6, 6))
pclass_counts.plot(
    kind="pie",
    autopct="%1.1f%%",
    labels=["1 класс", "2 класс", "3 класс"]
)
plt.title("Распределение пассажиров по классам на графике типа pie chart")
plt.ylabel("")
plt.show()
```

# Задание 6. Постройте график типа pairplot для всех числовых переменных датасета

```python
numeric_df = df.select_dtypes(include=["int64","float64"])
sns.pairplot(numeric_df)

plt.suptitle("График типа pairplot для всех числовых переменных", y=1)
plt.show()
```

# Задание 7. Постройте violin plot распределения возраста (Age) в зависимости от пола (Sex)

```python
plt.figure(figsize=(6, 4))

sns.violinplot(data=df, x="Sex", y="Age")

plt.title("Распределение возраста по полу")
plt.xlabel("Пол")
plt.ylabel("Возраст")

plt.show()
```

# Задание 8. Постройте heatmap корреляционной матрицы числовых признаков

```python
plt.figure(figsize=(30,10))

corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Heatmap корреляционной матрицы числовых признаков")

plt.show()
```

# Задание 9. Постройте интерактивный **sunburst plot** (plotly):
- первый уровень — количество пассажиров в каждом классе,
- второй уровень — количество женщин/мужчин в каждом классе  

```python
fig = px.sunburst(
    df,
    path=["Pclass", "Sex"],
    title="Количество пассажиров по классу и полу"
)

fig.update_traces(textinfo="label+percent entry+value")
fig.show()
```
